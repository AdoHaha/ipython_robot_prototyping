### Prototyping robots

Prototyping is producing a system that enables you to test your ideas. 
You want to create enough functionality to gather new **knowledge** that will guide your design. You don't want a "mini product" or a "crappy product", just to have enough to understand your issue better.

![](images/prototyping.jpg)



We will use Jupyter notebook in our robot prototyping because it enables us to rapidly create understandable program-chunks, visualize the incoming data and create simple control interfaces
